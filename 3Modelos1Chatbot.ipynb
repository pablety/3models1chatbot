```python
import torch
```

```python
def predLabel(new_text):

  # Tokeniza el nuevo texto
  tokens = tokenizer(new_text, truncation=True, padding='max_length', max_length=128)

  # Convierte los tokens en tensores de PyTorch
  input_ids = torch.tensor([tokens['input_ids']])
  attention_mask = torch.tensor([tokens['attention_mask']])

  # Pasa los tensores a través del modelo para hacer la predicción
  model.eval()
  with torch.no_grad():
      output = model(input_ids, attention_mask=attention_mask)

  # Obtiene las etiquetas predichas
  predicted_labels = torch.argmax(output.logits, dim=1)

  return predicted_labels.item()
```

```python
def predict_res(text):
    prep = predLabel(text)
    res_dict = {0: 'Clever',
                 1: 'CourtesyGoodBye',
                 2: 'CourtesyGreeting',
                 3: 'CourtesyGreetingResponse',
                 4: 'CurrentHumanQuery',
                 5: 'GoodBye',
                 6: 'Gossip',
                 7: 'Greeting',
                 8: 'GreetingResponse',
                 9: 'Jokes',
                 10: 'NameQuery',
                 11: 'NotTalking2U',
                 12: 'PodBayDoor',
                 13: 'PodBayDoorResponse',
                 14: 'RealNameQuery',
                 15: 'SelfAware',
                 16: 'Shutup',
                 17: 'Swearing',
                 18: 'Thanks',
                 19: 'TimeQuery',
                 20: 'UnderstandQuery',
                 21: 'WhoAmI'}
    return res_dict[prep]
```

```python
import json
pd.set_option("display.max_colwidth", None)
with open('/content/Intent.json') as f:
    data = json.load(f)

df = pd.DataFrame(data['intents'])


df_patterns = df[['text', 'intent']]


# Lista vacía para almacenar las filas del nuevo DataFrame
filas = []

# Iterar sobre el DataFrame original
for index, row in df_patterns.iterrows():
    lista = row['text']
    categoria = row['intent']

    # Crear una fila para cada elemento de la lista con la categoría correspondiente
    for elemento in lista:
        filas.append({'elemento': elemento, 'categoria': categoria})

# Crear el nuevo DataFrame
df_patrones = pd.DataFrame(filas)

# Imprimir el nuevo DataFrame
# print(df_patrones)

# df_patrones.categoria.value_counts()

# Agregar una columna con números únicos para cada categoría
df_patrones['codigo_categoria'] = pd.factorize(df_patrones['categoria'])[0]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_patrones.elemento, df_patrones.codigo_categoria, test_size=0.1)

from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()

X_train_cv = v.fit_transform(X_train.values)

X_train_np = X_train_cv.toarray()


from sklearn.naive_bayes import MultinomialNB

model_cat = MultinomialNB()
model_cat.fit(X_train_cv, y_train)

# emails = [
#     'Hey mohan, can we get together to watch footbal game tomorrow?',
#     'hello'
# ]

# emails_count = v.transform(emails)
# model_cat.predict(emails_count)
```

**GRADIO ChatBot cargando modelos segun codigos iniciales**

```python
import gradio as gr
import random
import time

with gr.Blocks() as demo:
    chatbot = gr.Chatbot()
    msg = gr.Textbox()
    clear = gr.ClearButton([msg, chatbot])
    print(chatbot)
    print(msg)
    # chat_history.append((message, bot_message))

    i = 0

    def respond(message, chat_history):
        # def identificar_prefijo(cadena):
        global state
        global tokenizer
        global model
        global i
        mensaje = ''
        if (i == 0 ):
          mensaje = '''Welcome, enjoy the chatbot experience!       '''
          i=1
        else :
          mensaje = ''
        if message.startswith("/1"):
                state = 1
                saludo = 'you are using gpt2'
                model = GPT2LMHeadModel.from_pretrained("/root/GPT2")
                tokenizer = GPT2Tokenizer.from_pretrained("/root/GPT2")

                prompt = message
                input_ids = tokenizer.encode(prompt, return_tensors="pt")
                output = model.generate(input_ids, max_length=50, num_return_sequences=1)
                generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

                bot_message =mensaje +  'you are using gpt2' + '___'+generated_text

        elif message.startswith("/2"):
                state = 2

                tokenizer = AutoTokenizer.from_pretrained("/root/BlenderBot")
                model = AutoModelForSeq2SeqLM.from_pretrained("/root/BlenderBot")
                # tokenizer = BlenderbotTokenizer.from_pretrained("/root/BlenderBot")
                # model = BlenderbotForConditionalGeneration.from_pretrained("/root/BlenderBot")
                utterance = message
                # Tokenize
                inputs = tokenizer(utterance, return_tensors="pt")
                # Passing through the utterances to the Blenderbot model
                res = model.generate(**inputs)
                # Decoding the model output
                bot_message = 'you are using blenderbot-ct'+'____'+tokenizer.decode(res[0])

        elif message.startswith("/3"):
                state = 3
                model = BertForSequenceClassification.from_pretrained('/root/BertCls')
                tokenizer = BertTokenizer.from_pretrained('/root/BertCls')

                bot_message = mensaje + "you are using bert-base-cased"+'___'+predict_res(message)


        else:
            if state == 1:

                prompt = message
                input_ids = tokenizer.encode(prompt, return_tensors="pt")
                output = model.generate(input_ids, max_length=50, num_return_sequences=1)
                generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

                bot_message = generated_text
                # return bot_message
            elif state == 2:

                utterance = message
                # Tokenize the utterance
                inputs = tokenizer(utterance, return_tensors="pt")
                # Passing through the utterances to the Blenderbot model
                res = model.generate(**inputs)
                # Decoding the model output
                tokenizer.decode(res[0])
                bot_message = tokenizer.decode(res[0])
                # return bot_message

            elif state == 3:

                bot_message = predict_res(message)

              # return bot_message
        chat_history.append((message, bot_message))
        time.sleep(5)
        return "", chat_history

    msg.submit(respond, [msg, chatbot], [msg, chatbot])
    print(msg)

demo.launch(debug=True)
```
